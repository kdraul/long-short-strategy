```python
import pandas as pd
```

```python
# imnporting list of tickers from a .csv file

tickers_list = pd.read_csv('../Assignment/tickers/tickers_list.csv')
tickers_list
```

```python
# Using random sampling to get a list 12% of the companies out of the total list of 750 companies
# i.e. a tickers list of approx. 90 companies.
# The tickers are picked from each index in proportion.

tickers = tickers_list.apply(lambda x : x.sample(frac = 0.12))
```

```python
tickers = pd.DataFrame(tickers.unstack().dropna())
tickers
```

```python
tickers = tickers.values.tolist()
tickers = [name for sub in tickers for name in sub]
tickers
```

```python
# importing yfinance library

import yfinance as yf
```

```python
# Downloading price data using yfinance

df = []

yahoo_tickers = [name + '.NS' for name in tickers]
yahoo_tickers

# Using 'Adj Close' column of the price data as the representation for stock price
for ticker in yahoo_tickers:
    tickdwld= yf.download(ticker, start = "2015-04-01", end = "2022-04-01")
    tickdwld = tickdwld['Adj Close']
    
    df.append(tickdwld)
```

```python
df = pd.DataFrame(df)
df= df.T
df.columns = tickers
df
```

```python
# Saving the tickers list and price data in '.csv' format for further analysis

tickers = pd.DataFrame(tickers)
tickers.to_csv('../Assignment/data/tickers.csv')
df.to_csv('../Assignment/data/price_data.csv')
```
